Vu que la mémoire saute si je veux lire tous les fichiers (2021-2023) d'un coup, je vais le faire séparement.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from tqdm import tqdm
```

# Données de 2021

```python
# Lire tous les fichiers de trafic
# A décommenter si l'on souhaite relire tous les fichiers
data_path = Path("data")
dic_trafic_dfs = {}

files = list(data_path.glob("*2021*.txt"))

for file in tqdm(files, desc="Chargement des fichiers"):
    df_name = f"df_{file.stem.lower()}"
    dic_trafic_dfs[df_name] = pd.read_csv(file, sep=";")

# Vérifier que les colonnes sont identiques dans tous les DFs
list_cols = list(dic_trafic_dfs.values())[0].columns.tolist()
for trafic_df in dic_trafic_dfs.values():
    if trafic_df.columns.tolist() != list_cols:
        print("les colonnes ne sont pas identiques dans tous les DFs")

# Concaténer tous les fichiers lus dans un même DataFrame
raw_trafic_2021_df = pd.concat(dic_trafic_dfs.values(), ignore_index=True)

# A décommenter si besoin de réécrire le fichier
raw_trafic_2021_df.to_parquet(
    "traffic_data_grand_paris_2021.parquet",
    engine="fastparquet",
    compression="snappy"
)
```
