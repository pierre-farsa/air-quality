```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from tqdm import tqdm
```

```python
dtype_map = {
    "iu_ac": "int32",
    "iu_nd_amont": "int32",
    "iu_nd_aval": "int32",
    "etat_trafic": "int8",
    "etat_barre": "int8",
    "q": "float32",
    "k": "float32",
    "libelle": "category"
}

trafic_2021_df = pd.read_parquet(
    "traffic_data_grand_paris_2021.parquet",
    columns=["iu_ac","libelle","iu_nd_amont","iu_nd_aval","t_1h","q","k","etat_trafic","etat_barre"]
).astype(dtype_map)

trafic_2021_df["t_1h"] = pd.to_datetime(trafic_2021_df["t_1h"])
```

```python
trafic_2021_df.dropna(subset=["q", "k"], inplace=True)
```
